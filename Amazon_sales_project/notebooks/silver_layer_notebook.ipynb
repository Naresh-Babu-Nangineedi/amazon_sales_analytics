```python
# Welcome to your new notebook
# Type here in the cell editor to add code!
from pyspark.sql.functions import col, when, lit, concat, split, to_date, year, trim, month, upper

# Read from Bronze
silver_df = spark.table("bronze_sales_raw")
```

```python
# Drop irrelevant/unnamed columns
silver_df = silver_df.drop("Unnamed_22")
```

```python
# Remove duplicate rows based on primary identifier (SKU + Order ID)
silver_df = silver_df.dropDuplicates(["SKU", "Order_ID"])
```

```python
# Replace all nulls in the DataFrame with "Unknown"
silver_df = silver_df.na.fill("Unknown")
```

```python
# Format string columns: remove whitespace, upper case where needed
silver_df = silver_df.withColumn("SKU", trim(upper(col("SKU")))) \
       .withColumn("Category", trim(upper(col("Category"))))

# Parse and enrich date fields
silver_df = silver_df.withColumn("OrderDate", to_date("Date", "MM-dd-yy")) \
       .withColumn("Month", month(col("OrderDate"))) \
       .withColumn("Year", year(col("OrderDate")))
```

```python
# Write to Silver Layer as a Delta Table
silver_df.write.mode("overwrite").saveAsTable("silver_sales_data")
```
