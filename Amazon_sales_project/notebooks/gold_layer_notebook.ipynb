```python
# Welcome to your new notebook
# Type here in the cell editor to add code!
from pyspark.sql.functions import col, to_date, month, year, dayofmonth, weekofyear, dayofweek

# Load Silver Layer table
df_silver = spark.table("silver_sales_data")
```

**Products Dimension Creation:**

```python
gold_dim_products = df_silver.select("SKU", "Category", "Style", "Size", "ASIN").dropDuplicates()
gold_dim_products.write.mode("overwrite").saveAsTable("gold_dim_products")
```

**Time dimension creation**

```python
gold_dim_time = df_silver.select("OrderDate") \
    .withColumn("Year", year("OrderDate")) \
    .withColumn("Month", month("OrderDate")) \
    .withColumn("Day", dayofmonth("OrderDate")) \
    .withColumn("Week", weekofyear("OrderDate")) \
    .withColumn("DayOfWeek", dayofweek("OrderDate")) \
    .dropDuplicates()

gold_dim_time.write.mode("overwrite").saveAsTable("gold_dim_time")
```

**Locations Dimension creation**

```python
gold_dim_locations = df_silver.select(
    "Ship_City", "Ship_State", "Ship_Postal_Code", "Ship_Country"
).dropDuplicates()

gold_dim_locations.write.mode("overwrite").saveAsTable("gold_dim_locations")
```

**Fact Table: Sales Creation**

```python
gold_fact_sales = df_silver.select(
    "Order_ID",
    "SKU",
    "Qty",
    "Amount",
    "Currency",
    "OrderDate",
    "Ship_City",
    "Ship_State",
    "Sales_Channel",
    "Promotion_IDs",
    "B2B"
)

gold_fact_sales.write.mode("overwrite").saveAsTable("gold_fact_sales")
```
